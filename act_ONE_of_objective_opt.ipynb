```python
Heating = data['Heating']
Cooling = data['Cooling']

features = data.drop(['Cooling','Heating'],axis=1)
```

```python
Clf_Heating = RandomForestRegressor()
Clf_Heating.fit(features,Heating)
y_Heating = Clf_Heating.predict(features)
print("r2 for Heating: {}".format(r2_score(y_Heating,Heating)))

Clf_Cooling = RandomForestRegressor()
Clf_Cooling.fit(features,Cooling)
y_Cooling = Clf_Cooling.predict(features)
print("r2 for Cooling: {}".format(r2_score(y_Cooling,Cooling)))
```

```python
flatint = np.zeros(len(features.columns))
for i ,k in enumerate(features.columns):
    if features[k].dtype !='float64':
        flatint[i] = 1
```

```python
flatint
```

```python
# first define your data and model
num = features.shape[1]
def fitness_function(solution):

    x = np.zeros(num)
    for i ,k in enumerate(features.columns):
        if features[k].dtype !='float64':
            x[i] = int(solution[i])
        else:
            x[i] = solution[i]
    # Make the predictions
    y_Heating = Clf_Heating.predict(x.reshape(1,-1))
    y_Cooling = Clf_Cooling.predict(x.reshape(1,-1))
    # Measure the performance
    average = (y_Heating * y_Cooling)/abs(y_Heating-y_Cooling)
    return average
    
maxx = np.zeros(num)
minn = np.zeros(num)

for j,k in enumerate(features.columns):
    maxx[j] = features[k].max()
    minn[j] = features[k].min()

problem = {
    "fit_func": fitness_function,
    "lb": list(minn),
    "ub": list(maxx),
    "minmax": "max",
}
```
